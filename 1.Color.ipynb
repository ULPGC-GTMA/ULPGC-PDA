Representamos colores. ¿Es buena la representación?

```python
white_xyY = colour.CCS_COLOURCHECKERS['ColorChecker24 - After November 2014'].data['white 9.5 (.05 D)']

col = 'green'
col_xyY = colour.CCS_COLOURCHECKERS['ColorChecker24 - After November 2014'].data[col]
col_XYZ = colour.xyY_to_XYZ(col_xyY)
col_sRGB = colour.XYZ_to_sRGB(col_XYZ)
print(col.title(), '\n xyY (cromaticidad): ', col_xyY, '\tsRGB: ', col_sRGB)


cplt.plot_single_colour_swatch(
    cplt.ColourSwatch(col_sRGB, col.title()),
    text_kwargs={'size': 'x-large'});

white_XYZ = colour.xyY_to_XYZ(white_xyY)
white_sRGB = colour.XYZ_to_sRGB(white_XYZ)
cplt.plot_single_colour_swatch(
    cplt.ColourSwatch(white_sRGB, 'Blanco'),
    text_kwargs={'size': 'x-large'});
```

Ahora hacemos el balance de blancos con un iluminante (D50)

```python
illuminant_xyY = colour.CCS_ILLUMINANTS['CIE 1931 2 Degree Standard Observer']['D50']

col = 'green'
col_xyY = colour.CCS_COLOURCHECKERS['ColorChecker24 - After November 2014'].data[col]
col_XYZ = colour.xyY_to_XYZ(col_xyY)
col_sRGB = colour.XYZ_to_sRGB(col_XYZ, illuminant = illuminant_xyY )
print(col.title(), '\n xyY (cromacidad): ', col_xyY, '\tsRGB: ', col_sRGB)


cplt.plot_single_colour_swatch(
    cplt.ColourSwatch(col_sRGB, col.title()),
    text_kwargs={'size': 'x-large'});

white_xyY = colour.CCS_COLOURCHECKERS['ColorChecker24 - After November 2014'].data['white 9.5 (.05 D)']
white_XYZ = colour.xyY_to_XYZ(white_xyY)
white_sRGB = colour.XYZ_to_sRGB(white_XYZ, illuminant = illuminant_xyY)
cplt.plot_single_colour_swatch(
    cplt.ColourSwatch(white_sRGB, 'Blanco'),
    text_kwargs={'size': 'x-large'});
```
