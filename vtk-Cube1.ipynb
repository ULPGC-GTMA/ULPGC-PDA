```python
import vtk

#import vtk.vtkInteractionStyle, vtk.vtkRenderingOpenGL2

from vtk.vtkCommonColor import vtkNamedColors
from vtk.vtkFiltersSources import vtkCubeSource

from vtk.vtkRenderingCore import (
    vtkActor,
    vtkPolyDataMapper,
    vtkRenderWindow,
    vtkRenderWindowInteractor,
    vtkRenderer
)
```

```python
colors = vtkNamedColors()
```

```python
# Create cube.
cube = vtkCubeSource()

# Forzamos actualización pues la fuente 
# es un proceso y no se corre si no se requiere
# En particular, no veríamos los puntos a la salida
cube.Update()

print(cube)
```

```python
cube_polydata = cube.GetOutput()

print('Clase: ', type(cube_polydata), '\n')
print(cube_polydata)
```

```python
points = cube_polydata.GetPoints()
print(type(points), '\n')
print(points)
```

```python
[points.GetPoint(pt) for pt in range(24)]
```

```python
# mapper
cubeMapper = vtkPolyDataMapper()
cubeMapper.SetInputData(cube.GetOutput())
```

```python
print(cubeMapper)
```

```python
# Actor.
cubeActor = vtkActor()
cubeActor.SetMapper(cubeMapper)
cubeActor.GetProperty().SetColor(colors.GetColor3d('Banana'))
```

```python
print(cubeActor)
```

```python
# Create a rendering window and renderer.
ren = vtkRenderer()
renWin = vtkRenderWindow()
renWin.SetWindowName('Cube1')
renWin.AddRenderer(ren)
```

```python
# Create a renderwindowinteractor.
iren = vtkRenderWindowInteractor()
iren.SetRenderWindow(renWin)
```

```python
# Assign actor to the renderer.
ren.AddActor(cubeActor)

ren.ResetCamera()
ren.GetActiveCamera().Azimuth(30)
ren.GetActiveCamera().Elevation(30)
ren.ResetCameraClippingRange()
ren.SetBackground(colors.GetColor3d('Silver'))
```

```python
# Enable user interface interactor.
#iren.Initialize()
#renWin.Render()
iren.Start()

del iren, renWin
```
